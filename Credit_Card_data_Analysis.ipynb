```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

# Settings
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (8, 5)
```

```python
# For CSV file
df = pd.read_csv("credit_card_data.csv")

# View first 5 rows
df.head()
```

```python
# Shape of dataset
df.shape

# Column names
df.columns

# Data types and null values
df.info()

# Statistical summary
df.describe()
```

```python
# Count of missing values
df.isnull().sum()

# Percentage of missing values
(df.isnull().sum() / len(df)) * 100
```

```python
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
# Check duplicates
df.duplicated().sum()

# Remove duplicates
df = df.drop_duplicates()
```

```python
df.hist(bins=30, figsize=(15, 10))
plt.show()
```

```python
for col in df.select_dtypes(include=np.number):
    sns.boxplot(x=df[col])
    plt.title(col)
    plt.show()
```

```python
for col in df.select_dtypes(include='object'):
    print(df[col].value_counts())
    print("-" * 40)
```

```python
sns.countplot(x=df['DAYS_BIRTH'])
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(x=df['AMT_INCOME_TOTAL'], y=df['AMT_CREDIT'])
plt.title('Income Total vs. Credit Amount')
plt.xlabel('Total Income')
plt.ylabel('Credit Amount')
plt.show()
```

```python
sns.boxplot(x='OCCUPATION_TYPE', y='AMT_INCOME_TOTAL', data=df)
plt.xticks(rotation=45)
plt.title('Income Total by Occupation Type')
plt.xlabel('Occupation Type')
plt.ylabel('Total Income')
plt.show()
```

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
for col in df.select_dtypes(include=np.number):
    sns.boxplot(x=df[col])
    plt.title(f"Outliers in {col}")
    plt.show()
```

```python
df.skew(numeric_only=True)
```

```python
sns.countplot(x=df['NAME_CONTRACT_TYPE'])
plt.title("Contract Type Distribution")
plt.show()
```

```python
sns.boxplot(x='NAME_CONTRACT_TYPE', y='AMT_CREDIT', data=df)
plt.title('Credit Amount by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Credit Amount')
plt.show()
```
