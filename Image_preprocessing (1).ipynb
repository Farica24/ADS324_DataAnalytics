```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

```python
# Load image (BGR format by default)
img = cv2.imread("/content/galaxy.jpg")

# Convert BGR to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.axis('off')
```

```python
img_resized = cv2.resize(img, (128, 128))
```

```python
img_array = np.array(img_resized)
print(img_array.shape)
```

```python
gray = cv2.cvtColor(img_resized, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap='gray')
plt.axis('off')
```

```python
normalized_img = gray / 255.0
```

```python
flattened_img = normalized_img.flatten()
print(flattened_img.shape)
```

```python
df = pd.DataFrame([flattened_img])
df.head()
```

```python
df['label'] = 1   # Example: 1 = positive class
```

```python
# Image Augmentation - Horizontal Flip
flip_h = cv2.flip(img, 1)
plt.imshow(flip_h)
plt.title("Horizontally Flipped Image")
plt.axis("off")
```

```python
# Image Augmentation - Vertical Flip
flip_v = cv2.flip(img, 0)
plt.imshow(flip_v)
plt.title("Vertically Flipped Image")
plt.axis("off")
```

```python
# Image Augmentation - Brightness Adjustment
bright = cv2.convertScaleAbs(img, alpha=1.2, beta=40)
plt.imshow(bright)
plt.title("Brightness Increased")
plt.axis("off")
```
